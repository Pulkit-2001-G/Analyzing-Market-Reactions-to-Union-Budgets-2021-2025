```python
#imported the required libraries

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# gathred the BSE index data from the Yahoo Finance.
bsesn = yf.Ticker('^BSESN')
bsesn_5y = bsesn.history(period = "5y")

# 5 year historic data is gathered.
# between 2021-01-01 to 2025-03-31.
```

```python
bsesn_5y.head() 

print(bsesn_5y)
# a preview of dataframe.
```

```python
bsesn_5y_a = bsesn_5y.reset_index()
# 'Date' is converted to column from the index.
```

```python
bsesn_5y_b = bsesn_5y_a.drop(['Dividends', 'Stock Splits'], axis = 1)

# columns 'Dividends' and 'Stock Splits' are dropped due to their insignificant data presence.

bsesn_5y_b.head()
```

```python
# converting the data type of 'Date' column of dataframe to 'DateTime'.
bsesn_5y_b['Date'] = pd.to_datetime(bsesn_5y_b['Date'])
```

```python
# filtering the data to fortnight before and after the decalration of union budget.

bsesn_1y = (bsesn_5y_b['Date'] >= '2021-01-18') & (bsesn_5y_b['Date'] <= '2021-02-15')
bsesn_2y = (bsesn_5y_b['Date'] >= '2022-01-18') & (bsesn_5y_b['Date'] <= '2022-02-15')
bsesn_3y = (bsesn_5y_b['Date'] >= '2023-01-18') & (bsesn_5y_b['Date'] <= '2023-02-15')
bsesn_4y = (bsesn_5y_b['Date'] >= '2024-01-18') & (bsesn_5y_b['Date'] <= '2024-02-15')
bsesn_5y = (bsesn_5y_b['Date'] >= '2025-01-18') & (bsesn_5y_b['Date'] <= '2025-02-15')

# a filter varaible is crated to filter the data based the the vale of the 'Date' column.

bsesn_2021_df = bsesn_5y_b[bsesn_1y]
bsesn_2022_df = bsesn_5y_b[bsesn_2y]
bsesn_2023_df = bsesn_5y_b[bsesn_3y]
bsesn_2024_df = bsesn_5y_b[bsesn_4y]
bsesn_2025_df = bsesn_5y_b[bsesn_5y]

# data is filtered and the new variable(dataframe) is created which contains the data in the specified time frame.
```

```python
# plotting the opeing and closing price of the BSE index prior and after the union budget announcement.

bsesn_2021_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2021 Opening & Closing Fortnight')
bsesn_2022_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2022 Opening & Closing Fortnight')
bsesn_2023_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2023 Opening & Closing Fortnight')
bsesn_2024_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2024 Opening & Closing Fortnight')
bsesn_2025_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2025 Opening & Closing Fortnight')
```

```python
# setps to export graphical representation for further use and analyses
# for the purpose of exporting the graphical representation the new variables is created named plot1, plot2 etc. respectedly.

plot11 = bsesn_2021_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2021 Opening & Closing Fortnight')
plot12 = bsesn_2022_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2022 Opening & Closing Fortnight')
plot13 = bsesn_2023_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2023 Opening & Closing Fortnight')
plot14 = bsesn_2024_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2024 Opening & Closing Fortnight')
plot15 = bsesn_2025_df.plot(x = 'Date', y = ['Close','Open'], title = 'BSE 2025 Opening & Closing Fortnight')

# a new variable is created figure(graph) to extract figure from the plot.

fig11 = plot11.get_figure()
fig12 = plot12.get_figure()
fig13 = plot13.get_figure()
fig14 = plot14.get_figure()
fig15 = plot15.get_figure()


# .savefig is utilised to save the figure as PNG.
fig11.savefig("BSE 2021 Fortnight", dpi=300, bbox_inches='tight')
fig12.savefig("BSE 2022 Fortnight", dpi=300, bbox_inches='tight')
fig13.savefig("BSE 2023 Fortnight", dpi=300, bbox_inches='tight')
fig14.savefig("BSE 2024 Fortnight", dpi=300, bbox_inches='tight')
fig15.savefig("BSE 2025 Fortnight", dpi=300, bbox_inches='tight')
```
