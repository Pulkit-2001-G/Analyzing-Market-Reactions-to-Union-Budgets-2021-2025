```python
#imported the required libraries

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# gathred the NSE index data from the Yahoo Finance.
nsei = yf.Ticker('^NSEI')
nsei_5y = nsei.history(period = "5y")

# 5 year historic data is gathered.
# between 2021-01-01 to 2025-03-31.
```

```python
nsei_5y.head() 

print(nsei_5y)
# a preview of dataframe.
```

```python
nsei_5y_a = nsei_5y.reset_index()
# 'Date' is converted to column from the index.
```

```python
nsei_5y_b = nsei_5y_a.drop(['Dividends', 'Stock Splits'], axis = 1)

# columns 'Dividends' and 'Stock Splits' are dropped due to their insignificant data presence.

nsei_5y_b.head()
```

```python
# converting the data type of 'Date' column of dataframe to 'DateTime'.
nsei_5y_b['Date'] = pd.to_datetime(nsei_5y_b['Date'])
```

```python
# filtering the data to five days before and after the decalration of union budget.

nesi_1y = (nsei_5y_b['Date'] >= '2021-01-26') & (nsei_5y_b['Date'] <= '2021-02-06')
nesi_2y = (nsei_5y_b['Date'] >= '2022-01-26') & (nsei_5y_b['Date'] <= '2022-02-06')
nesi_3y = (nsei_5y_b['Date'] >= '2023-01-26') & (nsei_5y_b['Date'] <= '2023-02-06')
nesi_4y = (nsei_5y_b['Date'] >= '2024-01-26') & (nsei_5y_b['Date'] <= '2024-02-06')
nesi_5y = (nsei_5y_b['Date'] >= '2025-01-26') & (nsei_5y_b['Date'] <= '2025-02-06')

# a filter varaible is crated to filter the data based the the vale of the 'Date' column.

nsei_2021_df = nsei_5y_b[nesi_1y]
nsei_2022_df = nsei_5y_b[nesi_2y]
nsei_2023_df = nsei_5y_b[nesi_3y]
nsei_2024_df = nsei_5y_b[nesi_4y]
nsei_2025_df = nsei_5y_b[nesi_5y]

# data is filtered and the new variable(dataframe) is created which contains the data in the specified time frame.
```

```python
# plotting the opeing and closing price of the NSE index prior and after the union budget announcement.

nsei_2021_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2021 Opening & Closing')
nsei_2022_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2022 Opening & Closing')
nsei_2023_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2023 Opening & Closing')
nsei_2024_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2024 Opening & Closing')
nsei_2025_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2025 Opening & Closing')
```

```python
# setps to export graphical representation for further use and analyses
# for the purpose of exporting the graphical representation the new variables is created named plot1, plot2 etc. respectedly.

plot1 = nsei_2021_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2021 Opening & Closing')
plot2 = nsei_2022_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2022 Opening & Closing')
plot3 = nsei_2023_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2023 Opening & Closing')
plot4 = nsei_2024_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2024 Opening & Closing')
plot5 = nsei_2025_df.plot(x = 'Date', y = ['Close','Open'], title = 'NSE 2025 Opening & Closing')

# a new variable is created figure(graph) to extract figure from the plot.

fig1 = plot1.get_figure()
fig2 = plot2.get_figure()
fig3 = plot3.get_figure()
fig4 = plot4.get_figure()
fig5 = plot5.get_figure()


# .savefig is utilised to save the figure as PNG.
fig1.savefig("NSE 2021 5d", dpi=300, bbox_inches='tight')
fig2.savefig("NSE 2022 5d", dpi=300, bbox_inches='tight')
fig3.savefig("NSE 2023 5d", dpi=300, bbox_inches='tight')
fig4.savefig("NSE 2024 5d", dpi=300, bbox_inches='tight')
fig5.savefig("NSE 2025 5d", dpi=300, bbox_inches='tight')
```
